```python
# using heatmaps to find null values
sns.set(rc={'figure.figsize':(8,10)})
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

### Merging the column

```python
df_country = pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_country
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.tail(2)
```

```python
# To check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
# Count the Countries
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
# pie chart - Top 3 countries that uses Zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

Observation:Zomato maximum records or transaction are from India After that USA and then United Kingdoms

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# Observation

1. Rating between 4.5 to 4.9 = Excellent
2. Rating between 4.0 to 4.4 = Very Good
3. Rating between 3.5 to 3.9 = Good
4. Rating between 2.5 to 3.4 = Average
5. Rating between 0.0 to 2.4 = Poor

```python
ratings.head()
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
# now according to colors
sns.set(rc={'figure.figsize':(12,6)})
sns.barplot(x="Aggregate rating",y="Rating Count", hue='Rating color' ,data=ratings, palette=['white','red','orange','yellow','green', 'green'])
```

## Observation

1. Not Rated count is very high
2. Maximum number of ratings are between 2.5 to 3.6

```python
# Count Plot
sns.countplot(x='Rating color', data=ratings, palette=['blue','red','orange','yellow','green', 'green'])
```
